1.Line plot

```python
import matplotlib.pyplot as plt

#sample data
months =['jan','feb','mar','april','may','june']
energy_consumption =[1200,1300,1100,1500,1400,1600]

#plot a line plt
plt.plot(months,energy_consumption,marker='o',linestyle='--',color='b')
plt.xlabel('Months')
plt.ylabel('Energy Consumption (MWh)')
plt.title('Energy Consumption In Six Months')
plt.show()
```

2.Bar Chart

```python
#Sample Data
energy_sources= ['Solar','Wind','Hydro Power','Bio Mass']
energy_Values = [1200,3400,2900,2500]
#creating bar chart
plt.bar(energy_sources,energy_Values,color='green')
# add title ,axis(lables)
plt.title("Energy Consumption by Renewable Energy sources")
plt.xlabel('Energy Sources')
plt.ylabel('Energy Consumption (MWh)')
plt.show()
```

Pie chart

```python
#Pie chart to show Energy Consumption Distribution
plt.pie(energy_Values,labels=energy_sources,autopct='%1.1f%%',colors=['yellow','blue','lightgreen','red'])
plt.title('Energy Consumption Share By Sources')
plt.show()
```

4.Scatter plot

```python
#sample data for energy consumption and carbon emissions
carbon_emissions = [200,300,500,450]# in kg CO2
#create a scatter plot
plt.scatter(energy_Values,carbon_emissions,color='b')
plt.title('Energy Consumption vs carbon Emission')
plt.xlabel('Energy Consumption in MWh')
plt.ylabel('Carbon Emission in kg CO2')
plt.show()
```

5.Customization

```python
#Create a bar chart for customization
plt.bar(energy_sources,energy_Values,color='orange',edgecolor='black',linewidth=2)
#add title and labels
plt.title('Customized Energy Consumption By source', fontsize=14,fontweight='bold')
plt.xlabel('Energy Source',fontsize=12)
plt.ylabel('Energy Consumption in MWh',fontsize=12)
#add grid andstyle
plt.grid(True,linestyle='--',alpha=0.7)
plt.legend(['Energy Consumption'],loc='upper left')
```
